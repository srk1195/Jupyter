<h4> += vs + (In-Place vs Not in-place) </h4>

```python
# In-place
a = np.array([1,2,3,4])
b = a
a += np.array([100,100,100,100])
# + symbol WITH the = operator modifies the existing array rather than creating a new one
print(b)
print(a)
```

```python
# not in-place
a = np.array([1,2,3,4])
b = a
a = a + np.array([100,100,100,100])
# + symbol WITHOUT the = operator creates a new array rather than modifing the existing array
print(b)
print(a)
```

```python
# not in-place
a = np.array([1,2,3,4,5])
slice_a = a[:3]
slice_a[0] = 100
print(slice_a)
a
```

```python
def variable_correlation(variable1, variable2):
    '''
    Fill in this function to calculate the number of data points for which
    the directions of variable1 and variable2 relative to the mean are the
    same, and the number of data points for which they are different.
    Direction here means whether each value is above or below its mean.
    
    You can classify cases where the value is equal to the mean for one or
    both variables however you like.
    
    Each argument will be a Pandas series.
    
    For example, if the inputs were pd.Series([1, 2, 3, 4]) and
    pd.Series([4, 5, 6, 7]), then the output would be (4, 0).
    This is because 1 and 4 are both below their means, 2 and 5 are both
    below, 3 and 6 are both above, and 4 and 7 are both above.
    
    On the other hand, if the inputs were pd.Series([1, 2, 3, 4]) and
    pd.Series([7, 6, 5, 4]), then the output would be (0, 4).
    This is because 1 is below its mean but 7 is above its mean, and
    so on.
    '''
    var1_mean = variable1.mean() # 72.869
    var2_mean = variable2.mean() # 9147.8799
    
    
    variable1_above = (variable1 > var1_mean)
    variable1_below = (variable1 < var1_mean)
    
    variable2_above = (variable2 > var2_mean)
    variable2_below = (variable2 < var2_mean)
    
    both_above = variable1_above & variable2_above
    both_below = variable1_below & variable2_below
    
    
    # print(both_above) # bool
    
    is_same_direction = both_above | both_below
    # print(is_same_direction)
    num_same_direction = is_same_direction.sum()        
    
    num_different_direction = len(is_same_direction) - num_same_direction
    
    return (num_same_direction, num_different_direction)

variable_correlation(life_expectancy,gdp)
```

<h5>Pearsons R</h5>

```python
def correlation(x, y):
    '''
    Fill in this function to compute the correlation between the two
    input variables. Each input is either a NumPy array or a Pandas
    Series.
    
    correlation = average of (x in standard units) times (y in standard units)
    
    Remember to pass the argument "ddof=0" to the Pandas std() function!
    '''
    # Standardizing the values
    x_std = ((x - x.mean()) / x.std(ddof=0))
    y_std = ((y - y.mean()) / y.std(ddof=0))
    
    # Calculate the Pearsons R ( -1 to +1 )
    pear_r = (x_std * y_std).mean()
    
    return pear_r

entries = nyc_riders_df['ENTRIESn_hourly']
cum_entries = nyc_riders_df['ENTRIESn']
rain = nyc_riders_df['meanprecipi']
temp = nyc_riders_df['meantempi']

print(correlation(entries, rain))
print(correlation(entries, temp))
print(correlation(rain, temp))
print(correlation(entries, cum_entries))
```

```python
grades_df = pd.DataFrame(
    data={'exam1': [43, 81, 78, 75, 89, 70, 91, 65, 98, 87],
          'exam2': [24, 63, 56, 56, 67, 51, 79, 46, 72, 60]},
    index=['Andre', 'Barry', 'Chris', 'Dan', 'Emilio', 
           'Fred', 'Greta', 'Humbert', 'Ivan', 'James']
)

def standardize_again(df):
    '''
    Fill in this function to standardize each column of the given
    DataFrame. To standardize a variable, convert each value to the
    number of standard deviations it is above or below the mean.
    
    This time, try to use vectorized operations instead of apply().
    You should get the same results as you did before.
    '''
    exam1_mean = df['exam1'].mean()
    exam2_mean = df['exam2'].mean()

    exam1_std = df['exam1'].std(ddof = 0)
    exam2_std = df['exam2'].std(ddof = 0)
    
    exam1_standardized = ( df['exam1'] - exam1_mean ) / exam1_std ;
    exam2_standardized = ( df['exam2'] - exam2_mean ) / exam2_std ;
    
    new_df = pd.DataFrame(
        data = {'exam1':exam1_standardized,
                'exam2':exam2_standardized
               }
    )
    
    return new_df

print(standardize(grades_df) == standardize_again(grades_df))
standardize_again(grades_df)
```
